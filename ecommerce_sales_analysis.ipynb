```python
import pandas as pd
```

```python
df = pd.read_csv("store_data.csv",encoding="latin1")
df.head()
```

```python
## Advanced Analysis
### Month- wise sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"],dayfirst=True)
monthly_sales = df.groupby(df["Order Date"].dt.month)["Sales"].sum()
monthly_sales.plot(kind="line",title="Monthly-Wise Sales Trend")
```

```python
df.shape
```

```python
loss_products = (df.groupby("Product Name")["Profit"].sum().sort_values().head(5))
loss_products
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
region_sales.plot(kind="bar",title="Region Sales")
```

```python
## Final Business Insights

#- Technology is the most profitable product category.
#- Some products consistently generate losses and should be reviewed.
#- Monthly sales show seasonality, with peaks in certain months.  
#- Certain  regions contribute more to total sales.
#- Discount levels have a direct impact on profit margin.    
```
